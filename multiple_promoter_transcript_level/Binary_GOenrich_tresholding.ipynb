```r
require(tidyverse)
require(data.table)
require(enrichR)
dbs <- listEnrichrDbs()
dbs%>% filter(grepl('KEGG_2016',libraryName) | grepl("GO_Biological_Process_2018",libraryName))
```

```r
go.res <- readRDS(file = "../dat/1901/enrichr.go.res.Rdata")
```

```r
ct<-'alpha1'
go.res[[ct]]$GO_Biological_Process_2018%>%top_n(-5,P.value)
```

```r
 system('open ../dat/1901/delta2.EnrichR.BP.100less.csv')
```

```r
for (ct in c("alpha1", "alpha2", "beta1", "beta2", "delta1", "delta2")) {
    fn = paste0("../dat/1901/", ct, ".EnrichR.BP.csv")
    fwrite(file = fn, x = go.res[[ct]]$GO_Biological_Process_2018 %>% separate(Overlap, 
            into = c("n.hit", "n.all"), sep = "/", convert = T)%>%arrange(P.value))}
```

```r
n = 100
for (ct in c("alpha1", "alpha2", "beta1", "beta2", "delta1", "delta2")) {
    fn = paste0("../dat/1901/", ct, ".EnrichR.BP.", n, "_less.csv")
    if (grepl("delta", ct)) {
        fwrite(file = fn, x = go.res[[ct]]$GO_Biological_Process_2018 %>% separate(Overlap, 
            into = c("n.hit", "n.all"), sep = "/", convert = T) %>% filter(n.all <= 
            n & P.value < 0.1) %>% arrange(Adjusted.P.value) %>% select(-one_of("Old.P.value", 
            "Old.Adjusted.P.value")))
        
    } else {
        fwrite(file = fn, x = go.res[[ct]]$GO_Biological_Process_2018 %>% separate(Overlap, 
            into = c("n.hit", "n.all"), sep = "/", convert = T) %>% filter(n.all <= 
            n & Adjusted.P.value < 0.1) %>% arrange(Adjusted.P.value) %>% select(-one_of("Old.P.value", 
            "Old.Adjusted.P.value", "Z.score")))
        
    }
    # system('open ../dat/1901/alpha2.EnrichR.BP.100less.csv')
}
```

```r
for(ct in c("alpha1","alpha2","beta1","beta2",'delta1','delta2')){
    fn = paste0("../dat/1901/",ct,".EnrichR.BP.150_less.csv")
    fwrite(file = fn,
       x = go.res[[ct]]$GO_Biological_Process_2018 %>% 
        separate(Overlap, into = c("n.hit","n.all"),sep = '/',convert =T)%>%
        filter(n.all<=150& Adjusted.P.value<0.1)%>%arrange(Adjusted.P.value)%>%
        select(-one_of("Old.P.value","Old.Adjusted.P.value","Z.score")))
#system("open ../dat/1901/alpha2.EnrichR.BP.100less.csv")
}
```

```r
for(ct in c("alpha1","alpha2","beta1","beta2")){
    fn = paste0("../dat/1901/",ct,".EnrichR.BP.all.csv")
    fwrite(file = fn,
       x = go.res[[ct]]$GO_Biological_Process_2018 %>% 
        separate(Overlap, into = c("n.hit","n.all"),sep = '/',convert =T)%>%
        arrange(Adjusted.P.value)%>%
        select(-one_of("Old.P.value","Old.Adjusted.P.value","Z.score")))
#system("open ../dat/1901/alpha2.EnrichR.BP.100less.csv")
}
```
